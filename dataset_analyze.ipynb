```python
from Bio import SeqIO
import pandas as pd
```

```python
# 1. 读取FASTA文件
sequences = []
for record in SeqIO.parse("D:/Coding_in_Python/deep_learning_porject_1/data/test1.fasta", "fasta"):
    # sequences.append(str(record.seq))
    sequences.append([str(record.description), str(record.seq)])

data=pd.DataFrame(sequences)
data.columns={'description', 'sequence'}
data
```

```python
data.info()
```

```python
data['length']=data.sequence.apply(lambda x:len(x))
data['toxicity']=data.description.str.split('|').str[1]
data
```

```python
data.toxicity=data.toxicity.astype(int)
data.toxicity=data.toxicity.map({1: 'Yes', 0: 'No'}).astype('category')
data
```

```python
data.to_csv('D:/Coding_in_Python/deep_learning_porject_1/data/test1_done.csv', index=False)
```
